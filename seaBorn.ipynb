importing the libraries


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

seaborn has some built in datasets

```python
tips=sns.load_dataset('tips')
tips.head()   ## head() --> it will give first five rows from datasets
```

```python
tips.tail()   ## tail() --> it will give last five row from datasets
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',
            col='time', hue='smoker', style='smoker')
### relplot is relative plot
```

load the iris dataset

```python
iris=sns.load_dataset('iris')
iris.head()
```

scatter plot of iris

```python
sns.scatterplot(x='sepal_length', y='petal_length',
                hue='species', data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

loading the titanic dataset


```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.tail()
```

counter plot

```python
sns.countplot(x='class',data=titanic)
```

```python
titanic.shape   ### it will give no of rows and columns
```

```python
sns.countplot(x='survived',data=titanic)  ## 0 represent person has not survived and 1 represent person has survived
```

bar chart

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

distribution plot

```python
sns.displot(titanic['age'].dropna())
```
