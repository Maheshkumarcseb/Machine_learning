Imbalanced dataset:

A dataset with unequal class distribution


```python
import numpy as np
import pandas as pd
```

Loading the dataset to pandas Dataframe

```python
credit_card_data = pd.read_csv('credit_data.csv')
```

```python
credit_card_data.shape
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
## distribution of the two classes
credit_card_data['Class'].value_counts()
```

1--> legit transaction
0-->fradulent transaction

```python
#seperating the legit and fradulent transaction
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

under-sampling

build a sample dataset containing similar distribution of legit and fraudulent tansaction

number of fradulent transaction--> 250

```python
legit_sample = legit.sample(n=250)
```

```python
print(legit_sample.shape)
```

Concatenate the two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
