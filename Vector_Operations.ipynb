```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,angles='xy',scale_units='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,angles='xy',scale_units='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,-3,-6,angles='xy',scale_units='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vector

```python
vector_2=np.array([0,0,2,3])
vector_3=np.array([0,0,3,-2])
vector_sum=vector_2+vector_3
print(vector_sum)
```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,angles='xy',scale_units='xy',scale=1,color='b')
plt.quiver(0,0,5,1,angles='xy',scale_units='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two vector

```python
vector_1=np.array([0,0,2,3])
vector_2=np.array([0,0,3,-2])
vector_sub=vector_1-vector_2
print(vector_sub)
```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,angles='xy',scale_units='xy',scale=1,color='b')
plt.quiver(0,0,-1,5,angles='xy',scale_units='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector with a scaler

```python
vector_1=np.array([0,0,2,3])
vector_2=vector_1*2
print(vector_2)
```

```python
# graph of multilpicaation of vector with scaler
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,4,6,angles='xy',scale_units='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1=np.array([0,0,2,3])
vector_2=-0.5*vector_1
print(vector_2)
#
```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,-1,-1.5,angles='xy',scale_units='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
#
```
